```python
import matplotlib.pyplot as plt
import pandas as pd

from gwp_aux import *
```

## Radiative Forcing from Global Mean Atmosphere with Stratospheric Temperature Adjustment.
From the supplemetary material from Shine and Myhre 2020 [1](https://agupubs.onlinelibrary.wiley.com/doi/epdf/10.1029/2019MS001951)

```python
df = pd.read_csv(
    "rf_gam_1cm_adjusted.csv",
    sep=";",
    decimal=",",
    engine="python"
)
```

```python
column_values = (
    df.iloc[:3000, 1]
    .astype(str)                            # ensure all entries are strings
    .str.replace(',', '.', regex=False)     # fix decimal separator
    .str.replace(' ', '', regex=False)      # remove stray spaces
)

RF_gam = pd.to_numeric(column_values, errors='coerce').to_numpy()

RF_gam *= 1e15
```

## Loading IR data from DFT calculation

```python
g631_b3_hr = '6-31G**/B3LYP_harm/orca.out'
g631_pw6_hr = "6-31G**/PW6B95_harm/orca.out"

jun_b3_hr = "jun-cc-pv/B3LYP_harm/orca.out"
jun_pw6_hr = "jun-cc-pv/PW6B95_harm/orca.out"

g6311_b3_hr = '6-311++G**/B3LYP_harm/orca.out'
g6311_pw_hr = '6-311++G**/PW6B95_harm/orca.out'

g6311_b3_an = '6-311++G**/B3LYP_anarm/orca.out'
g6311_pw6_an = "6-311++G**/PW6B95_anarm/orca.out"

orcas = [g631_b3_hr, g631_pw6_hr,jun_b3_hr, jun_pw6_hr, g6311_b3_hr, g6311_pw_hr, g6311_b3_an, g6311_pw6_an,]
```

## Radiative Efficiency and Global Warning Potential Calculation

```python
RE_ch4_ipcc = 0.000388

for o in orcas:
    #plotting
    r = plot_ir_spectrum(
        o,
        hwhm=30.0,
        scale_to_cm_molecule=True,
        save_plot=True,
        filename=o
    )

    #frequencies and intensities
    freqs = r[4]
    intensities = r[5]

    bin_centers, integrated_intensities = integrated_cross_section(intensities, freqs)

    # radiative efficiency and GWP
    RE = calculate_radiative_efficiency(integrated_intensities, RF_gam, tau=9.5, degradation_type='OH')
    GWP = calculate_gwp(RE, tau=9.5, molar_mass=16.04e-3, horizons=[20, 100, 500])

    # comparasion with IPCC reference
    abs_error = RE - RE_ch4_ipcc
    rel_error = (abs_error / RE_ch4_ipcc) * 100

    print(f"Treatment: {o}")
    print(f"Radiative efficiency (calc): {RE:.6f} W/m/ppb")
    print(f"IPCC CH RE: {RE_ch4_ipcc:.6f}")
    print(f"Abs error: {abs_error:.6e}, Rel error: {rel_error:.2f}%")
    print(f"GWP (20,100,500): {GWP}")
    print("-" * 50)
```
