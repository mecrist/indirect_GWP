## Radiative Forcing from Global Mean Atmosphere with Stratospheric Temperature Adjustment.
From the supplemetary material from Shine and Myhre 2020 [1](https://agupubs.onlinelibrary.wiley.com/doi/epdf/10.1029/2019MS001951)

```python
df = pd.read_csv(
    "rf_gam_1cm_adjusted.csv",
    sep=";",
    decimal=",",
    engine="python"
)
```

```python
column_values = (
    df.iloc[:3000, 1]
    .astype(str)                            # ensure all entries are strings
    .str.replace(',', '.', regex=False)     # fix decimal separator
    .str.replace(' ', '', regex=False)      # remove stray spaces
)

RF_gam = pd.to_numeric(column_values, errors='coerce').to_numpy()

RF_gam *= 1e15
```

```python
lista = list(range(1, len(RF_gam) + 1))

plt.figure(figsize=(12,6))
plt.plot(lista, RF_gam, color='red')
plt.xlabel(r'$cm^{-1}$')
plt.ylabel(r'$\frac{W}{m^2 \, ppb} \, (cm^{-1})^{-1} \, \left( \frac{cm^2}{\text{molcula}} \right)^{-1}$')
plt.title('Forante Radiativa da Atmosfera Global Anual Mdia')
plt.grid(True)
#plt.savefig('RF_GAM.png', dpi=800)
plt.show()
```

## Loading IR data from DFT calculation

```python
#ORDEM: BASE_FUNCIONAL_TRATAMENTO
g631_b3_an = "6-31G/B3LYP_anarm/orca.out"
g631_b3_hr = '6-31G/B3LYP_harm/orca.out'

#g631_pw6_an = "6-31G/PW6B95_anarm/orca.out" n rodou?
g631_pw6_hr = "6-31G/PW6B95_harm/orca.out"


jun_b3_an = "jun-cc-pv/B3LYP_anar/orca.out"
jun_b3_hr = "jun-cc-pv/B3LYP_harm/orca.out"

jun_pw6_an = "jun-cc-pv/PW6B95_anarm/orca.out"
jun_pw6_hr = "jun-cc-pv/PW6B95_harm/orca.out"

g6311_b3_an = '6-311++G%2A%2A/B3LYP_anarm/orca.out'
g6311_b3_hr = '6-311++G%2A%2A/B3LYP_harm/orca.out'

#g6311_pw_an = n rodou
g6311_pw_hr = '6-311++G%2A%2A/PW6B95_harm/orca.out'

orcas = [g631_b3_an, g631_b3_hr, g631_pw6_hr, jun_b3_an, jun_b3_hr, jun_pw6_an, jun_pw6_hr, g6311_b3_an, g6311_b3_hr, g6311_pw_hr]
```

## Radiative Efficiency and Global Warning Potential Calculation

```python
for o in orcas:
    r = plot_ir_spectrum(o, hwhm=30.0)
    freq = r[4] #frequency
    int = r[5] #intensity
    
    i_scaled = convert_kmpermol_to_cmpermolecule(int)

    avg_intensity = mean_cross_section(i_scaled, freq, delta_nu=1)
    
    RE = calculate_radiative_efficiency(avg_intensity, RF_gam, tau=11.8, degradation_type='OH')
    GWP = calculate_gwp(RE, tau=11.8, molar_mass=16.04, horizons=[20,100, 500])

    
    print(f"Treatment: {o}")
    print(f"Radiative efficiency: {RE}")
    print(f"Global Warning Potential: {GWP}")
```
